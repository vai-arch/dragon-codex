```python
# Quick analysis script
from pathlib import Path
from collections import Counter
from src.utils.config import Config

wiki_path = Config().WIKI_PATH
redirect_types = Counter()

for file in wiki_path.glob("*.txt"):
    with open(file, 'r', encoding='utf-8') as f:
        content = f.read()
        if "Naming_redirects" in content:
            redirect_types["Naming_redirects"] += 1
        elif "_redirects" in content.lower():
            # Find what type it is
            for line in content.split('\n'):
                if 'Categories:' in line or 'Category:' in line:
                    redirect_types[line.strip()] += 1
                    break

print("\nRedirect types found:")
for rtype, count in redirect_types.most_common():
    print(f"{rtype}: {count}")
```
