```python
# Check magic system index structure
import json
from src.utils.config import get_config

config = get_config()

with open(config.FILE_MAGIC_SYSTEM_INDEX, 'r', encoding='utf-8') as f:
    magic_index = json.load(f)

print(f"Total magic concepts: {len(magic_index)}")
print("\nFirst 10 magic concepts:")
for i, concept in enumerate(list(magic_index.keys())[:10]):
    category = magic_index[concept].get('category', 'Unknown')
    print(f"  {i+1}. {concept} ({category})")

print("\nExample magic concept structure:")
first_concept = list(magic_index.keys())[0]
print(json.dumps({first_concept: magic_index[first_concept]}, indent=2)[:500])
```
